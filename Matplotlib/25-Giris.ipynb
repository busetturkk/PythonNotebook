```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList=[10,20,30,30,30,40,50,60,70,75]
```

```python
weightList=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyAgeList=np.array(ageList)
numpyWeightList=np.array(weightList)
```

```python
numpyAgeList
```

```python
numpyWeightList
```

```python
plt.plot(numpyAgeList,numpyWeightList,"b")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişimi")
#plt.show()
```

```python
numpyArray1=np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2=numpyArray1**3
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1,numpyArray2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.subplot(1,2,2) #1 sıra,2 kolon,şu an 2.grafiği çizdiriyoruz
plt.plot(numpyArray2,numpyArray1,"g--")

```

```python
myFigure=plt.figure()

figureAxes=myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyArray1,numpyArray2,"g")
figureAxes.set_xlabel("X ekseni Veri İsmi")
figureAxes.set_ylabel("y ekseni veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2=plt.figure()

axis1=figure2.add_axes([0.1,0.1,0.9,0.9])
axis2=figure2.add_axes([0.2,0.6,0.3,0.3])

axis1.plot(numpyArray1,numpyArray2,"g")
axis1.set_xlabel("X ekseni")
axis1.set_ylabel("Y ekseni")
axis1.set_title("Ana Grafik Başlık")

axis2.plot(numpyArray2,numpyArray1,"b")
#axis2.set_xlabel("X ekseni")
#axis2.set_ylabel("Y ekseni")
axis2.set_title("Küçük Grafik Başlık")

```

```python
(figure, myAxes)=plt.subplots(nrows=1,ncols=2)
for axis in myAxes:
    axis.plot(numpyArray1,numpyArray2,"g")
    axis.set_xlabel("X ekseni")
plt.tight_layout()
```

```python
newFigure=plt.figure(dpi=100)
#dpi=100

newAxis=newFigure.add_axes([0.1,0.1,0.9,0.9])

newAxis.plot(numpyArray1,numpyArray1**2, label="numpyArray**2")
newAxis.plot(numpyArray1,numpyArray1**3, label="numpyArray**3")
newAxis.legend(loc=2)
```

```python
newFigure.savefig("myFigure.png",dpi=200)
```
