```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyArray1= np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2=numpyArray1**2
```

```python
(myFig,myAxis)=plt.subplots()
myAxis.plot(numpyArray1,numpyArray2,color="#C96F23",alpha=0.5)
myAxis.plot(numpyArray2,numpyArray1,"b")
```

```python
(newFig, newAxis)= plt.subplots()
newAxis.plot(numpyArray1,numpyArray1+2, color="blue", linewidth=5.0)
newAxis.plot(numpyArray1, numpyArray1+3, color="pink", linewidth=4.0)

newAxis.plot(numpyArray1, numpyArray1+4, color="#C965BD", linestyle="-.")
newAxis.plot(numpyArray1, numpyArray1+5, color="#C965BD", linestyle="--")
newAxis.plot(numpyArray1, numpyArray1+6, color="#C965BD", linestyle=":")

newAxis.plot(numpyArray1, numpyArray1+7, color="#000000", linestyle="--", marker="o",markersize=8, markerfacecolor="red")
newAxis.plot(numpyArray1, numpyArray1+8, color="#000000", linestyle="--", marker="+",markersize=8)
```
